# Configs

```python
def font_config(
        size: int
        ) -> dict:

    temp_config = dict(
        family = 'Arial',
        size = size,
        color = 'white',
        )

    return temp_config


x_axis_config = dict(
        # showline = False,
        # showgrid = True,
        # showticklabels = True,
        # linecolor = 'white',
        # linewidth = 1,
        # ticks = 'outside',
        tickfont = font_config(size = 16)
        )

y_axis_config = dict(
        #showline = False,
        #showgrid = True,
        #showticklabels = True,
        #linecolor = 'white',
        #linewidth = 1,
        #ticks = 'outside',
        tickfont = font_config(size = 16)
        )

margin_config = dict(
        autoexpand = True,
        l = 100,
        r = 20,
        t = 80,
        b = 20
        )

legend_config = dict(
        font = font_config(18)
        )
```

# Heatmaps

```python
flights: pd.DataFrame = sns.load_dataset('flights')
flights.head()
```

```python
fig = px.density_heatmap(
        data_frame = flights,
        x='year',
        y = 'month',
        z = 'passengers',
        #color_continuous_scale = 'Viridis',
        marginal_x = 'histogram',
        marginal_y = 'histogram',
        text_auto = True
        )

fig = fig.update_layout(
        #autosize = True,

        height = 800,
        width = 1400,

        yaxis_title = 'MONTH',

        xaxis_title = 'YEAR',
        xaxis = dict(side = 'bottom')
        )

fig.show()
```
